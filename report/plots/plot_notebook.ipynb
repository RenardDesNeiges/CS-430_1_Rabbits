# Plots for report 1 of the CS-430 exercices

```python
import csv
import sys
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# File parsing routine

def parseCSV(address):
        """ Parses the txt file (comma-separated ==> CSV) that we use to export the plot from RePast
        """

        with open(address, newline='', encoding="utf-8-sig") as csvfile:

            reader = csv.reader(csvfile, delimiter=',', quotechar='|')

            first = True
            index = []
            tick = []
            grassTiles = []
            rabbits = []
            
            for row in reader:
                if first:
                    index = row
                    print(index)
                    first = False
                else:
                    tick.append(float(row[0]))
                    grassTiles.append(float(row[1]))
                    rabbits.append(float(row[2]))
            
            return (np.array(tick),np.array(grassTiles),np.array(rabbits))
```

```python
# Plotting Routine

def plotPop(ticks,grass,rabbits,start,length,name,maxAxis = -1):
    
    #defining the figure style
    plt.style.use('seaborn-whitegrid')
    
    end = start + length
    
    #initialize the plot object
    fig = plt.figure()
    ax = plt.axes()
    
    #setting the axis limits
    plt.xlim(-10+ticks[start], ticks[end]+10)
    if maxAxis == -1:
        plt.ylim(0, max([max(grass[start:end]),max(rabbits[start:end])])+10);
    else:
        plt.ylim(0, maxAxis);
    
    #plotting the grass
    x = ticks[start:end]
    y = grass[start:end]
    ax.plot(x, y,color="green", linewidth=1.0)
    
    #plotting the rabbits
    x = ticks[start:end]
    y = rabbits[start:end]
    ax.plot(x,y,color="red", linewidth=1.0)

    plt.savefig(name+".png",dpi=300)
```

```python
# creating a list of dictonaries for simulation parameter and properties 

simulations = []
```
