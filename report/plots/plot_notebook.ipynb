```python
import csv
import sys
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# File parsing routine

def parseCSV(address):
        """ Parses the txt file (comma-separated ==> CSV) that we use to export the plot from RePast
        """

        with open(address, newline='', encoding="utf-8-sig") as csvfile:

            reader = csv.reader(csvfile, delimiter=',', quotechar='|')

            first = True
            index = []
            tick = []
            grassTiles = []
            rabbits = []
            
            for row in reader:
                if first:
                    index = row
                    print(index)
                    first = False
                else:
                    tick.append(float(row[0]))
                    grassTiles.append(float(row[1]))
                    rabbits.append(float(row[2]))
            
            return (np.array(tick),np.array(grassTiles),np.array(rabbits))
```

```python
# Plotting Routine

def plotPop(ticks,grass,rabbits,length,name):
    
    #defining the figure style
    plt.style.use('seaborn-whitegrid')
    
    #initialize the plot object
    fig = plt.figure()
    ax = plt.axes()
    
    #setting the axis limits
    plt.xlim(-10, ticks[length]+10)
    plt.ylim(0, max([max(grass[0:length]),max(rabbits[0:length])])+10);
    
    #plotting the grass
    x = ticks[0:length]
    y = grass[0:length]
    ax.plot(x, y,color="green", linewidth=1.0)
    
    #plotting the rabbits
    x = ticks[0:length]
    y = rabbits[0:length]
    ax.plot(x,y,color="red", linewidth=1.0)

    plt.savefig(name+".png",dpi=300)
```

```python
# creating a list of dictonaries for simulation parameter and properties 

simulations = []
```

## Simulation Run 1 (reference run) :
* 29 sept. 2020 à 11:51:16
* RngSeed: 1601373075162
* BirthTreshold: 20
* GrassEnergy: 5.0
* GrassGrowthRate: 50
* GridSize: 20
* NumInitGrass: 100
* NumInitRabbits: 30

```python
(ticks_1,grass_1,rabbits_1) = parseCSV('simRun_1.txt')
```

```python
# plotting the reference simulation run
plotPop(ticks_1,grass_1,rabbits_1,400,"sim1")
```

```python
#basic quantifiers of the first run

start = 200
frame = 500
stop = start + frame

stable_grass_1 = grass_1[start:stop]
stable_rabbits_1 = rabbits_1[start:stop]

print("variance of grass # = " + str(np.var(stable_grass_1)))
print("variance of rabbits # = " + str(np.var(stable_rabbits_1)))
print("average of grass # = " + str(np.average(stable_grass_1)))
print("average of rabbits # = " + str(np.average(stable_rabbits_1)))

simParams = {
    "BirthTreshold": 20,
    "GrassEnergy": 5.0,
    "GrassGrowthRate": 50,
    "GridSize": 20,
    "NumInitGrass": 100,
    "NumInitRabbits": 30,
    "VarGrass": np.var(stable_grass_1),
    "VarRabbits": np.var(stable_rabbits_1),
    "AvgGrass": np.average(stable_grass_1),
    "AvgRabbits": np.average(stable_rabbits_1)
}

simulations.append(simParams)
```
